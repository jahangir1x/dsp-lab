```python
import numpy as np
import matplotlib.pyplot as plt
```

<center> y = 5sin(2 * pi * 500t + 90) </center>

```python
def plot_custom_sine_wave(start, stop, sampling_rate):
    x = np.linspace(start = start, stop = stop, num = int((stop-start) * sampling_rate)) # + time_shift
    y = 5 * np.sin(2 * np.pi * 300 * x + np.deg2rad(90))

    plt.title('Custom Sine Wave')
    plt.grid(True)
    plt.plot(x, y) # analog signal
    # plt.stem(x, y) # digital signal
    plt.xlabel("time")
    plt.ylabel("amplitude")
    plt.show()

plot_custom_sine_wave(start=0, stop=0.004, sampling_rate=1000)
```
