```python
import numpy as np
import matplotlib.pyplot as plt
```

<center> Sine Wave </center>

```python
def sine_wave(amplitude, phase_deg):
    x = np.linspace(start = -100, stop = 100, num = 100)
    y = amplitude * np.sin(2 * np.pi * x + np.deg2rad(phase_deg))

    min_y = y.min()
    max_y = y.max()
    quantization_bit = 8
    max_level = 2 ** quantization_bit - 1
    quantized_y = np.round(max_level * (y - min_y) / (max_y - min_y))

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    ax1.set_title('Original Signal')
    ax1.grid(True)
    ax1.plot(x, y)
    ax1.set_xlabel("time")
    ax1.set_ylabel("amplitude")
    
    ax2.set_title('Quantized Signal')
    ax2.grid(True)
    ax2.step(x, quantized_y)
    ax2.set_xlabel("time")
    ax2.set_ylabel("amplitude")

    plt.show()

sine_wave(amplitude=1, phase_deg=0)
```
