```python
import numpy as np
import matplotlib.pyplot as plt


def convolution(x, h):
    len_x = len(x)
    len_h = len(h)
    len_y = len_x + len_h - 1
    y = np.zeros(len_y)

    for n in range(len_y):
        for k in range(len_x):
            if 0 <= n - k < len_h:
                y[n] += x[k] * h[n - k]

    return y


h = np.array([1, 2, 3, 4])
x = np.array([2, 5, 7, 8])

y_np_convolve = np.convolve(x, h)
y_custom_convolve = convolution(x, h)

n_x = np.arange(len(x))
n_h = np.arange(len(h))
n_y = np.arange(len(y_np_convolve))

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

ax1.stem(n_x, x, linefmt="b-", markerfmt="bo", basefmt=" ", label="x(n)")
ax1.stem(n_h, h, linefmt="r-", markerfmt="ro", basefmt=" ", label="h(n)")
ax1.stem(
    n_y,
    y_np_convolve,
    linefmt="g-",
    markerfmt="go",
    basefmt=" ",
    label="y(n) = x(n) * h(n) (np.convolve)",
)
ax1.set_title("Convolution Result using np.convolve()")
ax1.set_xlabel("n")
ax1.set_ylabel("Amplitude")
ax1.grid(True)
ax1.legend()

ax2.stem(n_x, x, linefmt="b-", markerfmt="bo", basefmt=" ", label="x(n)")
ax2.stem(n_h, h, linefmt="r-", markerfmt="ro", basefmt=" ", label="h(n)")
ax2.stem(
    n_y,
    y_custom_convolve,
    linefmt="g-",
    markerfmt="go",
    basefmt=" ",
    label="y(n) = x(n) * h(n) (custom)",
)
ax2.set_title("Convolution Result using custom convolution()")
ax2.set_xlabel("n")
ax2.set_ylabel("Amplitude")
ax2.grid(True)
ax2.legend()

plt.show()
```
