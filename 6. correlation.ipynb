# DFT for book example


```python
import numpy as np
import matplotlib.pyplot as plt


def custom_correlation(x, y):
    N = len(x)
    M = len(y)
    # Pad y with zeros to perform full correlation
    y_padded = np.pad(y, (N - 1, N - 1), "constant")
    correlation = []
    for k in range(N + M - 1):
        sum_val = 0
        for n in range(N):
            sum_val += x[n] * y_padded[n + k]
        correlation.append(sum_val)
    return np.array(correlation)


x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

correlation = np.correlate(x, y, mode="full")
lags = np.arange(-len(x) + 1, len(y))
plt.subplot(3, 2, 1)
plt.stem(lags, correlation, basefmt=" ")
plt.title("Cross-Correlation of x(n) and y(n)")
plt.xlabel("Lag")
plt.ylabel("Correlation")
plt.grid(True)

correlation = custom_correlation(x, y)
lags = np.arange(-len(x) + 1, len(y))
plt.subplot(3, 2, 2)
plt.stem(lags, correlation, basefmt=" ")
plt.title("Cross-Correlation (custom) of x(n) and y(n)")
plt.xlabel("Lag")
plt.ylabel("Correlation")
plt.grid(True)

plt.show()
```
