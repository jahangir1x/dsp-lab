```python
import numpy as np
import matplotlib.pyplot as plt

frequencies = np.linspace(0, 1000, 1000)
omega = 2 * np.pi * frequencies

numerator_coeffs = [1, 2, 2]
denominator_coeffs = [1, 3, 5]

s = 1j * omega
numerator = np.polyval(numerator_coeffs, s)
denominator = np.polyval(denominator_coeffs, s)
H = numerator / denominator

magnitude_response = np.abs(H)
phase_response = np.angle(H)

plt.figure(figsize=(14, 6))

plt.subplot(2, 1, 1)
plt.plot(frequencies, magnitude_response)
plt.title("Magnitude Response of H(s)")
plt.xlabel("Frequency [Hz]")
plt.ylabel("|H(jω)|")
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(frequencies, np.degrees(phase_response))
plt.title("Phase Response of H(s)")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Phase [degrees]")
plt.grid(True)

plt.tight_layout()
plt.show()
```
