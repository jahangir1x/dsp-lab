```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def convolve(x, h):
    len_x = len(x)
    len_h = len(h)
    len_y = len_x + len_h - 1
    y = [0] * len_y
    for i in range(len_y):
        for k in range(len_x):
            if(i - k >= 0 and i - k < len_h):
                y[i] += x[k] * h[i - k]
    return y
```

```python
x = [1, 2, 3, 4, 5]
h = [2, 3, 4]
h_folded = h[::-1]
y = convolve(x, h)
y_corr = convolve(x, h_folded) # Correlation is same as Convolution except the folding
                               # Correlation does not require folding

plt.subplot(2, 3, 1)
plt.stem(range(len(x)), x)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('x(n)')
plt.grid()

plt.subplot(2, 3, 2)
plt.stem(range(len(h)), h)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('h(n)')
plt.grid()

plt.subplot(2, 3, 3)
plt.stem(range(len(y)), y)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('y(n) = x(n) * h(n)')

plt.subplot(2, 3, 4)
plt.stem(range(len(x)), x)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('x(n)')
plt.grid()

plt.subplot(2, 3, 5)
plt.stem(range(len(h)), h)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('h(n)')
plt.grid()

plt.subplot(2, 3, 6)
plt.stem(range(len(y_corr)), y_corr)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('y(n) = x(n) . h(n)')
plt.tight_layout()
plt.show()
# print(y)
```
